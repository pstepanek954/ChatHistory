```python
import pandas as pd
import time
import datetime


df = pd.read_csv("./timeseries2.csv",encoding="utf-8")

# df_1 = df[df["messageType"].isin([47])]
# df_1 = df_1[df_1["mesDes"].isin([1])]
# df_1
# list_t 指的是结构化了之后的time
list_t = list(df["msgCreateTime"])
for i in range(len(list_t)):
    list_t[i] =  time.localtime(list_t[i])

map_total_msg = {}
map_month_msg = {}


for i, content in df.iterrows():
    current_date = time.strftime("%Y-%m-%d", list_t[i])
    current_month = time.strftime("%Y-%m", list_t[i])
    
    if current_date not in map_total_msg:
        map_total_msg[current_date] = [0, 0, 0, 0, 0, 0]
    if current_month not in map_month_msg:
        map_month_msg[current_month] = [0, 0, 0, 0, 0, 0]
    if content["mesDes"] == 1:
        map_total_msg[current_date][0] += 1
        map_month_msg[current_month][0] += 1
    else:
        map_total_msg[current_date][1] += 1
        map_month_msg[current_month][1] += 1
    if content["messageType"] == 47:
        if content["mesDes"] == 1:
            map_total_msg[current_date][3] += 1
            map_month_msg[current_month][3] += 1
        else:
            map_total_msg[current_date][4] += 1
            map_month_msg[current_month][4] += 1
        map_total_msg[current_date][5] += 1
        map_month_msg[current_month][5] += 1
        
    map_total_msg[current_date][2] += 1
    map_month_msg[current_month][2] += 1
    
map_total_msg
```

```python
import datetime
# 用来返回所有时间
def get_full_time(path):
    df = pd.read_csv(path)
    time_all = list(df["msgCreateTime"])
    list_t_2 = []
    for i in time_all:
        timeStamp = i
        dateArray = datetime.datetime.fromtimestamp(timeStamp)
        otherStyleTime = dateArray.strftime("%Y-%m-%d %H:%M:%S")
        list_t_2.append(otherStyleTime)
    return list_t_2

def add_nan_time(map_4_time):
    begin = datetime.date(2021,10,8)
    end = datetime.date(2022,3 ,18)
    for i in range((end - begin).days + 1 ):
        temp = str(begin + datetime.timedelta(days=i) )
        if temp not in map_4_time:
            map_4_time[temp] = [0,0,0, 0, 0, 0]
    return map_4_time
```

```python
from pyecharts.charts import WordCloud, Tab
def word_cloud(s_name, title_, data):
    c = WordCloud()
    c.add(series_name=s_name, data_pair=data, word_size_range=[20, 80],rotate_step="10",shape= "star",
)
    c.set_global_opts(
        title_opts=opts.TitleOpts(
            title=title_, pos_left = "center", title_textstyle_opts=opts.TextStyleOpts(font_size=25,font_family="Times New Roman")
        ),
        tooltip_opts=opts.TooltipOpts(is_show=True),
    )
    return c
tab = Tab()
tab.add(word_cloud("聊天分词形容词", "形容词",res_adj_), "形容词词云")
tab.add(word_cloud("聊天分词名词","名词",res_n_), "名词词云")
tab.add(word_cloud("聊天分词动词","动词",res_v_), "动词词云")
# tab.add(grid_mutil_yaxis(), "grid-example")
tab.render("./toyty/tab_wc_base.html")
```
